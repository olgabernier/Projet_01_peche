```python
import numpy as np

a = np.loadtxt('peche_clean_sans_titre.csv',dtype = np.int32, delimiter = ",")
a
```

```python
#4. Quels sont les mois où l’on vend, en moyenne, le plus de coquilles Saint-Jacques ?

def decoupe(t):
    '''Retourne les arrays correspondant aux colonnes du tableau'''
    lst = []
    for i in range(t.shape[1]):
        lst.append(t[:,i])
    lst2 = np.array(lst)
    return lst2
        

# Nommer les colonnes:
annee, mois, ventes, valeur, invendu = decoupe(a)


month = np.unique(mois)
avg_sales = np.array( [ventes[mois ==x].mean() for x in month])
avg_sales = list(zip(month,avg_sales))
avg_sales.sort(key = lambda x:x[1])
avg_sales.reverse()


def avg_sales_per_month():
    '''Retourne les ventes moyennes mensuelles par ordre décroissant'''
    month = np.unique(mois)
    avg_sales = np.array( [ventes[mois ==x].mean() for x in month])
    avg_sales = list(zip(month,avg_sales))
    avg_sales.sort(key = lambda x:x[1])
    avg_sales.reverse() 
    return avg_sales


def avg_sales_months_GGG():
    '''Retourne les ventes moyennes mensuelles par ordre décroissant'''
    month = np.unique(mois)
    avg_sales = np.array( [ventes[mois ==x].mean() for x in month])
    avg_sales = list(zip(month,avg_sales))
    avg_sales.sort(key = lambda x:x[1])
    avg_sales.reverse() 


def avg_sales_per_month_graph():
    import matplotlib.pyplot as plt
    month = np.unique(mois)
    avg_sales = np.array( [ventes[mois ==x].mean() for x in month])
    x = avg_sales
    plt.pie(x,labels = np.unique(mois))
    plt.ylabel('Month')
    plt.title('Average sales per month, kg')
   
```
